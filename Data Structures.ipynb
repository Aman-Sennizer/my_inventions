```python
class linkedlist:
    class node:
        __slots__='_element','_next'
        def __init__(self,element,next):
            self._element=element
            self._next=next
            
    def __init__(self):
        self.head=None
        self.tail=None
        self._size=0
    
    def __len__(self):
        return self._size
    
    
    def is_empty(self):
        return self._size==0
        
    def insert_at_beginning(self,e):
        new=self.node(e,None)
        if self.is_empty():
            self.head=new
            self.tail=new
        else:
            new._next=self.head
        self.head=new
        self._size=self._size+1
        
    def insert_at_end(self,e):
        new=self.node(e,None)
        if self.is_empty():
            self.head=new
            self.tail=new
        else:
            self.tail._next=new
        self.tail=new
        self._size=self._size+1
        
    def insert_at_position(self,e,pos):
        new=self.node(e,None)
        i=1
        thead=self.head
        while i<pos-1:
            thead=thead._next
            i=i+1
        new._next=thead._next
        thead._next=new
        self._size=self._size+1
        
    def display(self):
        thead=self.head
        while thead:
            print(thead._element,end='--->')
            thead=thead._next
            
    def delete_at_beginning(self):
        if self.is_empty():
            print("Not possible")
        else:
            self.head=self.head._next
            self._size-=1
    
    def delete_at_end(self):
        if self.is_empty():
            print("Not possible")
        else:
            thead=self.head
            while thead._next!=None:
                ptr=thead
                thead=thead._next
            self.tail=ptr
            ptr._next=None
            self._size=self._size-1
        
    def delete_at_position(self,pos):
        i=1
        thead=self.head
        while i<pos-1:
            thead=thead._next
            i=i+1
        thead._next=thead._next._next
        self._size=self._size-1  
        
l=linkedlist()
l.insert_at_beginning(5)
l.insert_at_end(8)
l.insert_at_beginning(5)
l.insert_at_beginning(5)
l.insert_at_end(10)
l.insert_at_beginning(5)
l.delete_at_beginning()
l.delete_at_end()
l.delete_at_end()
l.insert_at_end(10)
l.insert_at_end(15)
l.insert_at_position(20,4)
l.delete_at_position(4)
l.display()
```

```python
class Node:
    def __init__(self,value):
            self.value=value
            self.left=None
            self.right=None
    
    
def insert(root,node):
    
    if node.value<root.value:
        
        if(root.left==None):
            root.left=node
        else:
            insert(root.left,node)
    
    if node.value>root.value:
        if(root.right==None):
            root.right=node
        else:
            insert(root.right,node)
        
    
    return node
                
def preorder(root):
    if root:
        print(root.value)
        preorder(root.left)
        preorder(root.right)

def inorder(root):
    if root:
        inorder(root.left)
        print(root.value)
        inorder(root.right)
def postorder(root):
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.value)
    
root=Node(50)
insert(root,Node(10))
insert(root,Node(2))
insert(root,Node(80))
insert(root,Node(15))
insert(root,Node(60))
insert(root,Node(90))
#preorder(root)
#inorder(root)
postorder(root)
def min_value(root):
    while root.left!=None:
        root=root.left
    print(root.value)
min_value(root)

def max_value(root):
    while root.right!=None:
        root=root.right
    print(root.value)
max_value(root)
```
